Checked 

```python
district_df = df.groupby(
    ['state', 'district']
).agg(
    total_updates=('total_updates', 'sum'),
    adult_updates=('bio_age_17_', 'sum'),
    child_updates=('bio_age_5_17', 'sum')
).reset_index()
```

```python
district_df['adult_share_pct'] = (
    district_df['adult_updates'] / district_df['total_updates']
) * 100
```

```python
district_df.head()
```

Data was aggregated at the district level to assess geographic disparities in adult biometric update participation.

```python
state_df = district_df.groupby('state').agg(
    total_updates=('total_updates', 'sum'),
    adult_updates=('adult_updates', 'sum'),
    child_updates=('child_updates', 'sum')
).reset_index()

state_df['adult_share_pct'] = (
    state_df['adult_updates'] / state_df['total_updates']
) * 100
```

```python
national_adult_avg = district_df['adult_share_pct'].mean()
national_adult_avg
```

The national average adult participation share will be used as a benchmark for comparative analysis

```python
district_df['below_national_avg'] = (district_df['adult_share_pct'] < national_adult_avg)
```

```python
district_df['below_national_avg'].value_counts()
```

```python
district_df.sort_values('adult_share_pct').head(10)
```

Top 10 Lowest Adult Participation Districts

```python
state_df.sort_values('adult_share_pct').plot(
    x='state', y='adult_share_pct',
    kind='bar', figsize=(12,4)
)

plt.axhline(national_adult_avg, linestyle='--')
plt.title("Adult Share of Biometric Updates by State")
plt.ylabel("Adult Share (%)")
plt.xlabel("State")
plt.show()
```

Substantial inter-state variation is observed in adult biometric update participation.

```python
district_df['adult_share_pct'].plot(kind='hist', bins=20)
plt.title("Distribution of Adult Share Across Districts")
plt.xlabel("Adult Share (%)")
plt.show()
```

The distribution  shows long tails at both extremes .A subset of districts records very low adult participation, including values approaching zero, which may signal significant access barriers, mobility limitations, or service delivery gaps for adult populations in these areas.

```python
monthly_df = df.groupby('year_month').agg(
    total_updates=('total_updates', 'sum'),
    adult_updates=('bio_age_17_', 'sum'),
    child_updates=('bio_age_5_17', 'sum')
).reset_index()

monthly_df['adult_share_pct'] = (
    monthly_df['adult_updates'] / monthly_df['total_updates']
) * 100
```

```python
monthly_df.head()
```

Daily data is usually noisy. So i did Monthly aggregated which shows real trends

```python
plt.figure(figsize=(10,4))
plt.plot(
    monthly_df['year_month'].astype(str),
    monthly_df['adult_share_pct'],
    marker='o'
)

plt.xticks(rotation=45)
plt.title("National Trend of Adult Biometric Update Participation")
plt.ylabel("Adult Share (%)")
plt.xlabel("Month")
plt.show()
```

```python
low_states = (
    state_df.sort_values('adult_share_pct')
    .head(3)['state']
    .tolist()
)

low_states
```

This states and ut has the lowest participation in adult biometric updates

```python
state_month_df = df.groupby(
    ['state', 'year_month']
).agg(
    total_updates=('total_updates', 'sum'),
    adult_updates=('bio_age_17_', 'sum')
).reset_index()

state_month_df['adult_share_pct'] = (
    state_month_df['adult_updates'] /
    state_month_df['total_updates']
) * 100
```

```python
plt.figure(figsize=(10,4))

for st in low_states:
    subset = state_month_df[state_month_df['state'] == st]
    plt.plot(
        subset['year_month'].astype(str),
        subset['adult_share_pct'],
        marker='o',
        label=st
    )

plt.legend()
plt.xticks(rotation=45)
plt.title("Adult Participation Trends in Low-Performing States")
plt.ylabel("Adult Share (%)")
plt.xlabel("Month")
plt.show()
```

```python
state_month_df['below_national_avg'] = (
    state_month_df['adult_share_pct'] < national_adult_avg
)

persistence_score = (
    state_month_df.groupby('state')['below_national_avg']
    .mean()
    .sort_values(ascending=False)
)

persistence_score.head(5)
```

1.Uttar Pradesh & Mizoram (1.0): These states never exceeded the national average during the observed period. They are persistently below the benchmark.
2.Madhya Pradesh, Arunachal Pradesh, & Assam (0.888): These states were below the national average during the observed period which is 9 months

```python
def risk_label(adult_share):
    if adult_share < 30:
        return 'High Risk'
    elif adult_share < 50:
        return 'Medium Risk'
    else:
        return 'Low Risk'

district_df['risk_category'] = district_df['adult_share_pct'].apply(risk_label)
```

```python
district_df['risk_category'].value_counts()
```

Risk Categorization Framework: Districts were classified into High, Medium, and Low risk categories based on adult biometric update participation. Low adult participation is treated as a proxy for potential exclusion due to mobility, biometric failure, or access barriers.

```python
district_df['risk_category'].value_counts().plot(kind='bar')
plt.title("Distribution of Districts by Exclusion Risk Category")
plt.ylabel("Number of Districts")
plt.show()
```
